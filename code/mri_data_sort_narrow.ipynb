## Narrowing down the dataset by various criteria

```python
import pandas as pd
from dateutil import parser
import numpy as np
```

```python
loadPath = '../data/'
writePath = '../data/'
```

Loading and extracting data.

```python
# Load the CSV file
df = pd.read_csv(loadPath + 'MRIT1_longitudinal.csv')

# all the unique patients
patients = df['NACCID'].unique()

# number of MRIT1 scans each unique patients
number_of_scans = len(df['NACCID'])

# Count the occurrences of each patient (NACCID) in the dataset
naccid_counts = df['NACCID'].value_counts()
```

Calculating time interval between scans.

```python
differences = []
sum = 0                                                     # sum to store indexing of the patients (borders for what dates to read)

for i in range(0, len(patients)):                               # one subarray for each unique NACCID (patient)

    naccid = df['NACCID'].unique()[i]                           # unique patients

    num = naccid_counts.loc['{}'.format(naccid)]                # number of MRIT1 scans


    months = np.array(df['MRIMO'][sum : sum + num])
    day = np.array(df['MRIDY'][sum : sum + num])
    years = np.array(df['MRIYR'][sum : sum + num])


    diffs = [0]


    date1 = np.zeros(num, dtype=object)


    for k in range(0, num):
        
        date1[k] = ('{}-{}-{}'.format(years[k], months[k], day[k]))
        
        date1[k] = parser.parse(date1[k])

        
    for l in range(0, len(date1)-1, 1):
            
        d = np.abs(date1[l+1] - date1[l])
        

        diffs.append(d.days)

    sum += num

    differences.append(diffs)
```

```python
# transforming the difference to months
for p in differences:

    for e in range(0, len(p)):

        p[e] = p[e] / 30.437  
```

Saving the data to a dictionary.

```python
dic_unsorted = {}

for i in range(0, len(patients)):    
    
    dic_unsorted['{}'.format(patients[i])] = differences[i] 


print(dic_unsorted)
```

Sorting the dictionary by lenghts of values and last element in values.

```python
# sorting the dictionary by lenghts of value array and by last element in array
sorted_items = sorted(dic_unsorted.items(), key=lambda x: (len(x[1]), (x[1][-1])))          

# Rebuilding the dictionary manually
dic_sorted = {k: v for k, v in sorted_items}
```

```python
print(dic_sorted)
```

### Narrowing the data 
I will delete each patient that had multiple scans taken on the same day - that data is not longitudinal. 

```python
naccid_trash = [k for k, v in dic_sorted.items() if v.count(0.0) > 1]
```

```python
print(naccid_trash)
print(len(naccid_trash))             # we get rid of 166 patients
```

Removing patients that are in the naccid_trash array.

```python
for i in naccid_trash:
    dic_sorted.pop('{}'.format(i))
```

```python
print(dic_sorted)
print(len(dic_sorted))                # 1778 - 166 = 1612
```

Writing down the narrowed down patients in a new .csv file.

```python
df = pd.read_csv(loadPath + 'MRIT1_longitudinal.csv')
```

```python
# extracting keys in sorted dictionary
keys = dic_sorted.keys()
```

```python
df_narrow = df[df['NACCID'].isin(keys)]

print(len(df_narrow))
```

```python
# Select only the relevant columns and rename them according to the new file structure
df_final = df_narrow[['NACCID', 'NACCMNUM', 'MRIMO', 'MRIDY', 'MRIYR', 'NACCMRIA', 'NACCMRFI']]
df_final.columns = ['NACCID', 'NACCMNUM', 'MRIMO', 'MRIDY', 'MRIYR', 'NACCMRIA', 'NACCMRFI']

# Check the final dataset before saving
print("Final dataset preview:")
print(df_final.head())
```

```python
# sanity check
print(len(df_narrow['NACCID'].unique()))
```

### Narrowing by MRIT1 scan within 1 year of UDS visits
I will keep only the patients, that had their MRIT1 taken within 1 year of UDS visit. That way I can match results from scans and their diagnosis.

```python
import os
```

I will read all the files in within1yr directory, which is a subset of scans taken within one year of a UDS visit.

```python
dirPath = '../../NACC_data/within1yr/nifti'
```

```python
# lists all the filenames in specified directory
zip_array = os.listdir(dirPath)
```

```python
print(zip_array)
```

Problem is, that the names in zip_array have "ni" at the end, while filenames in .csv files don't. That's why I did not find any matches at first. So I manually remove all the "ni" and then compare.

```python
zip_array = [file_name.replace('ni', '') for file_name in zip_array]
```

```python
print(zip_array)
```

Getting the filenames stored in MRT1_longitudinal_narrow_v1.csv.

```python
df = pd.read_csv(loadPath + 'MRIT1_longitudinal_narrow_v1.csv')
```

```python
naccid_filenames = df['NACCMRFI']
```

Finding common files.

```python
common_filenames = set(zip_array).intersection(set(naccid_filenames))

print(f"Number of common NACCIDs: {len(common_filenames)}")
```

We have 2785 MRIT1 scans within 1 year of UDS visit. Now I will check how many unique patients that is. 

```python
df = df[df['NACCMRFI'].isin(common_filenames)]
```

```python
patients = df['NACCID'].unique()
```

```python
print(len(patients))
```

In this file, so far, we went from 1778 patients to 1612 to 1377. 

Writing the narrowed down data to a MRIT1_longitudinal_narrow_v2.csv file

```python
df_final = df[['NACCID', 'NACCMNUM', 'MRIMO', 'MRIDY', 'MRIYR', 'NACCMRIA', 'NACCMRFI']]
df_final.columns = ['NACCID', 'NACCMNUM', 'MRIMO', 'MRIDY', 'MRIYR', 'NACCMRIA', 'NACCMRFI']
```

### Narrowing the UDS .csv file data
Before I head into matching the UDSD with MRIT1 longitudinal data, I need to narrow down tha patients in UDS .csv file the same I did it for longitudinal data. <br>
<br>
This is easily achievable with comparing the NACCIDS in MRIT1_longitudinal_narrow_v2.csv and MRIT1_longitudinal_UDS.csv

```python
df1 = pd.read_csv(loadPath + 'MRIT1_longitudinal_narrow_v2.csv')
df2 = pd.read_csv(loadPath + 'MRIT1_longitudinal_UDS.csv')
```

```python
patients1 = df1['NACCID'].unique()
patients2 = df2['NACCID'].unique()
```

```python
common_naccids = set(patients1).intersection(set(patients2))
```

```python
df2 = df2[df2['NACCID'].isin(common_naccids)]
```

```python
df2_final = df2[['NACCID', 'NACCVNUM', 'VISITMO', 'VISITDAY', 'VISITYR', 'NACCUDSD', 'NACCALZD']]
df2_final.columns = ['NACCID', 'NACCVNUM', 'VISITMO', 'VISITDAY', 'VISITYR', 'NACCUDSD', 'NACCALZD']
```

```python
df2_final.to_csv(writePath + 'MRIT1_longitudinal_UDS_narrow_v2.csv', index=False)

print("Data writing to CSV complete!")
```

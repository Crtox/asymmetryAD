```python
# Convert the relative path to an absolute path
ncPath = os.path.abspath(ncPath)
mciPath = os.path.abspath(mciPath)
alzdPath = os.path.abspath(alzdPath)
transPath = os.path.abspath(transPath)

# Modify the absolute path for long path support on Windows
if os.name == 'nt':                     # Check if the operating system is Windows
    ncPath = '\\\\?\\' + ncPath
    mciPath = '\\\\?\\' + mciPath
    alzdPath = '\\\\?\\' + alzdPath
    transPath = '\\\\?\\' + transPath
```

```python
pattern = 'NACC'
```

```python
# List to store the matching folders
matching_folders = []

# Iterate over the items in the directory (change Path names to access different cohorts)
for item in os.listdir(ncPath):
    item_path = os.path.join(ncPath, item)
    
    # Check if the item is a folder and matches the pattern
    if os.path.isdir(item_path) and re.match(pattern, item):
        matching_folders.append(item_path)
```

```python
delete_empty = []
```

```python
for folder in matching_folders:

    subfolder = os.listdir(folder)
    subfolder_fp = os.path.join(folder, subfolder[0])

    subs = os.listdir(subfolder_fp)

    if len(subs) == 0: 

        delete_empty.append(folder)
```

### Deleting files in folders 
Now i will read newly corrected .csv files: nc_v3_2.csv, mci_v3.csv, alzd_v3.csv, trans_v3.csv. I will compare files stored in those files with files in my cohorts and delete any file in cohort that is not found in my .csv files.

```python
nc_df = pd.read_csv(dataPath + 'nc_v3_2.csv')
mci_df = pd.read_csv(dataPath + 'mci_v3.csv')
alz_df = pd.read_csv(dataPath + 'alzd_v3.csv')
trans_df = pd.read_csv(dataPath + 'trans_v3.csv')
```

```python
nc_filenames = list(nc_df['NACCMRFI'])
mci_filenames = list(mci_df['NACCMRFI'])
alz_filenames = list(alz_df['NACCMRFI'])
trans_filenames = list(trans_df['NACCMRFI'])
```

```python
print(nc_filenames)
print(len(nc_filenames))
print(len(mci_filenames))
print(len(alz_filenames))
print(len(trans_filenames))
```

```python
nc_dir_filenames = os.listdir(ncPath)
mci_dir_filenames = os.listdir(mciPath)
alz_dir_filenames = os.listdir(alzdPath)
trans_dir_filenames = os.listdir(transPath)
```

```python
print(nc_dir_filenames)
print(len(nc_dir_filenames))
print(len(mci_dir_filenames))
print(len(alz_dir_filenames))
print(len(trans_dir_filenames))
```

```python
nc_common_filenames = set(nc_filenames).intersection(set(nc_dir_filenames))
mci_common_filenames = set(mci_filenames).intersection(set(mci_dir_filenames))
alz_common_filenames = set(alz_filenames).intersection(set(alz_dir_filenames))
trans_common_filenames = set(trans_filenames).intersection(set(trans_dir_filenames))
```

```python
print(len(nc_common_filenames))
print(len(mci_common_filenames))
print(len(alz_common_filenames))
print(len(trans_common_filenames))
```

```python
nc_delete = []
mci_delete = []
alz_delete = []
trans_delete = []
```

```python
for i, j, k, l in itertools.zip_longest(nc_dir_filenames, mci_dir_filenames, alz_dir_filenames, trans_dir_filenames):

    if i is not None and i not in nc_common_filenames:
        nc_delete.append(i)

    if j is not None and j not in mci_common_filenames:
        mci_delete.append(j)

    if k is not None and k not in alz_common_filenames:
        alz_delete.append(k)

    if l is not None and l not in trans_common_filenames:
        trans_delete.append(l)
```

```python
print(len(nc_delete))
print(len(mci_delete))
print(len(alz_delete))
print(len(trans_delete))
```
